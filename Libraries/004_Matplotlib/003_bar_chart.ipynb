# Bar Chart

###### A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. It can be created using the bar() method.
> _Syntax:_ 
_plt.bar(x, height, width, bottom, align)_

```python

import matplotlib.pyplot as plt 
import numpy as np
```

```python
# data to display on plots 
x = [3, 1, 6, 5, 2, 7, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 
  
# This will plot a simple bar chart
plt.bar(x, y)
  
# Title to the plot
plt.title("Bar Chart")
  
# Adding the legends
plt.legend(["bar"])
plt.show()
```

```python
product_code = np.array([1,2,3,4,5])
sale_price = np.array([13, 25, 37, 48, 65])
profit = np.array([2, 3, 4, 2, 8])

plt.bar(product_code-0.05, sale_price, label = "Price", color='blue')
plt.bar(product_code, profit, label = "Prodit", color='green')
plt.legend(loc="best", shadow=True)
plt.show
```
