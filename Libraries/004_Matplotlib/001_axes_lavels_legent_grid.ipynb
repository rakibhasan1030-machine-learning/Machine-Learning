# Axes Lavels, Legent and Grid

```python
import matplotlib.pyplot as plt
```

```python
days = [1, 2, 3, 4, 5]
max = [10, 3, 7, 13, 2]
min = [8, 5, 2, 12, 15]
avg = [20, 8, 11, 6, 13]

plt.plot(days, max, label='Max')
plt.plot(days, min, label='Min')
plt.plot(days, avg, label='Avg')

# Adding the title
plt.title("Weather Plot")
  
# Adding the labels
plt.xlabel("Days")
plt.ylabel("Weather Condition")
plt.legend(loc="best", shadow=True)
plt.grid()
plt.show
```
