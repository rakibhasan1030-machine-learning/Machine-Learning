### Logistic Regression - IRIS DATASET

> A regression algirthm which does classification

## Problem
> #### Train a logistic regression classifire to predict wheather a flower is Iris-Virginica or not!


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris = datasets.load_iris()
```

```python
# print output for understanding
def printOutput(predicted_ans):
    if predicted_ans == 0:
        print(f'Predicted answer is {predicted_ans}, that means it\'s not a "Iris-Virginica"')
    else:
        print(f'Predicted answer is {predicted_ans}, that means it\'s a "Iris-Virginica"')
```

```python
# load iris dataset
iris = datasets.load_iris()

# print descriptoins of the datasets
# print(f"Dataset Keys : {iris.keys()}")
# print(f"Data shape (ROW x COLUMN) : {iris['data'].shape}") 
# print(iris.DESCR)

# we take patel width as feature which is in index 3 and slice it from main dataset
x = iris['data'][:, 3:]
# print(X)

# as we work only with iris virginica flower based on question, 
# so we convert targets as label in binery classifire
# if the target value is equal 2 then it will convert into TRUE or 1
# otherwise it will convert into FALSE or 0
# Y = iris['target']
# print(Y)

# astype(int) from NUMPY - convert True to 1 and false to 0
y = (iris['target'] == 2).astype(int)
# print(Y)



# train a logistic regression classifier 
clf = LogisticRegression()
clf.fit(x, y)
predict = clf.predict([[2.8]])
print(printOutput(predict))

# using matplotlib to plot the visualization
x_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_prob = clf.predict_proba(x_new)
plt.plot(x_new, y_prob[:, 1], "g-")

# Adding the title
plt.title("Iris-Virginica")
plt.grid()
plt.show()
```
